```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Chargement des donn√©s

data = pd.read_csv('sortie.txt').to_numpy()
Procs = []
Times = []
for i in range(data.size):
    d = data[i][0].split(" ")
    #print(d)
    Procs.append(float(d[0]))
    Times.append(float(d[1]))

print(Procs)
print(np.round(Times,2))
```

```python
plt.plot(Procs, np.round(Times,2))
plt.xlabel("Processeurs")
plt.ylabel("Temps")
plt.legend()
```
